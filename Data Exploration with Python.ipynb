```python
df.query('Location != "Other"') \
.groupby('Location')['speed_mph'] \
.agg(['mean','count']) \
.query ('count >=10') \
.sort_values('mean')
```

```python
ax = df.query('Location != "Other"') \
.groupby('Location')['speed_mph'] \
.agg(['mean','count']) \
.query ('count >=10') \
.sort_values('mean')['mean'] \
.plot(kind = 'barh', figsize = (12,5) , title = "Coaster Location by Average Speed")
ax.set_xlabel("Avg. Coaster Speed")
plt.show()
```
