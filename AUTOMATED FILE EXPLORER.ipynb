```python
###CORRECTED CODE

import os
import shutil

path = r"/Users/oluwakemiayoola/Downloads/"
file_names = os.listdir(path)
folder_names = ['image files', 'picture files']

# Create folders if they don't exist
for folder_name in folder_names:
    if not os.path.exists(os.path.join(path, folder_name)):
        os.makedirs(os.path.join(path, folder_name))

# Move files to the respective folders
for file in file_names:
    if ".png" in file and not os.path.exists(os.path.join(path, 'image files', file)):
        shutil.move(os.path.join(path, file), os.path.join(path, 'image files', file))

    elif ".jpg" in file and not os.path.exists(os.path.join(path, 'picture files', file)):
        shutil.move(os.path.join(path, file), os.path.join(path, 'picture files', file))
    elif ".avif" in file and not os.path.exists(os.path.join(path, 'picture files', file)):
        shutil.move(os.path.join(path, file), os.path.join(path, 'picture files', file))
```
